```python
Supervised Learning

Supervised learning is a machine learning approach where algorithms learn from labeled
data, meaning they are trained on examples with known inputs and corresponding outputs,
enabling them to make predictions or classifications on new, unseen data. 
Here&#39;s a more detailed explanation:
Key Concepts:
 Labeled Data:
Supervised learning algorithms are trained using datasets where each data point has a known
&quot;label&quot; or &quot;output&quot; value. 
 Prediction and Classification:
The goal of supervised learning is to train a model that can accurately predict the output for
new, unseen data points based on the patterns learned from the labeled training data. 
 Types of Supervised Learning:
 Regression: Predicts continuous values (e.g., predicting house prices based on
features). 
 Classification: Predicts discrete categories or classes (e.g., classifying emails
as spam or not spam). 

 Examples of Supervised Learning Algorithms:
 Linear Regression: Predicts a continuous target variable by finding the best-
fit line or plane. 
 Logistic Regression: Predicts the probability of an event occurring (e.g.,
predicting whether a customer will click on an ad). 
 Support Vector Machines (SVM): Used for both classification and
regression, finding the optimal hyperplane to separate data points. 
 K-Nearest Neighbors (KNN): Classifies data points based on the majority
class of their nearest neighbors. 
 Naive Bayes: A probabilistic classifier based on Bayes&#39; theorem, often used
for text classification. 
 Decision Trees: Creates a tree-like model to make decisions based on
features. 
 Random Forests: An ensemble method that combines multiple decision trees
for improved accuracy. 
 Neural Networks: Complex models that learn from data through
interconnected nodes, used for various tasks like image recognition and
natural language processing. 

How it Works:
1. Data Collection and Preparation:
Gather labeled data, ensuring it is clean, relevant, and representative of the problem you want
to solve.
2. Algorithm Selection:
Choose a suitable supervised learning algorithm based on the problem type (regression or
classification) and the characteristics of the data. 
3. Model Training:
Feed the labeled data to the chosen algorithm, allowing it to learn the relationships between
the inputs and outputs. 
4. Model Evaluation:
Assess the performance of the trained model using metrics like accuracy, precision, recall,
and F1-score. 
5. Model Tuning:
Adjust the algorithm&#39;s parameters to optimize its performance on the evaluation data. 
6. Prediction:
Use the trained model to make predictions or classifications on new, unseen data. 
Applications of Supervised Learning:
 Predictive Analytics: Forecasting trends, sales, and other business metrics.
 Image and Object Recognition: Identifying objects and scenes in images and videos.
 Spam Detection: Filtering out unwanted emails.
 Fraud Detection: Identifying suspicious transactions.
 Customer Segmentation: Grouping customers based on their behavior and
characteristics.
 Medical Diagnosis: Assisting doctors in diagnosing diseases.
 Natural Language Processing: Tasks like sentiment analysis and machine
translation.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pd.DataFrame({'Age':[36,42,23,52,43,44,66,35,52,35,24,18,45],
'Experience':[10,12,4,4,21,14,3,14,13,5,3,3,9],
'Rank':[9,4,6,4,8,5,7,9,7,9,5,7,9],
'Nationality':['UK','USA','N','USA','USA','UK','N','UK','N','N','USA','UK','UK'],
'Go':['NO','NO','NO','NO','YES','NO','YES','YES','YES','YES','NO','YES','YES']
})
print(df)

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']
X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()

dtree = dtree.fit(X,y)

tree.plot_tree(dtree, feature_names=features)
plt.show()
```
